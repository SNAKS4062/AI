```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.preprocessing import LabelEncoder
```

```python
labels = ['_'.join(os.path.basename(img_path).split('_')[:-1:]) for img_path in image_names]

label_encoder = LabelEncoder()

label_encoder.fit(labels)

features = []
labels_encoded = []
IMAGE_SIZE = (224, 224)

for name in image_names:
    label = '_'.join(name.split('_')[:-1:])
    
    label_encoded = label_encoder.transform([label])[0]

    if label_encoded is not None:
        img = load_img(os.path.join(image_fp, name))
        img = tf.image.resize_with_pad(img_to_array(img, dtype='uint8'), *IMAGE_SIZE).numpy().astype('uint8')
        image = np.array(img)
        features.append(image)
        labels_encoded.append(label_encoded) 
```

```python
features_array=np.array(features)
labels_array=np.array(labels)
```

```python
features_array
```

```python
labels_one_hot=pd.get_dummies(labels_array, dtype='int')
```

```python
labels_one_hot
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(features_array, labels_one_hot, test_size=0.2, random_state=42)
```

```python
X_train, X_val, y_train, y_val=train_test_split(X_train, y_train, test_size=0.25)
```

```python
from tensorflow.keras import layers,Input,Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.mobilenet import preprocess_input as pp_i
from tensorflow.keras.layers import RandomFlip, RandomRotation, Dense, Dropout
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
```

```python
num_classes = labels_one_hot.shape[1]
```

```python
data_augmentation = Sequential ([RandomFlip ("horizontal_and_vertical"), RandomRotation(0.2)])
prediction_layers = Dense (num_classes, activation = 'softmax')
```

```python
resnet_model = ResNet50(include_top = False, pooling = 'avg', weights = 'imagenet')
resnet_model.trainable = False
preprocess_input = pp_i
```

```python
inputs = Input(shape=(224,224,3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = resnet_model(x, training=False)
x = Dropout(0.2)(x)
outputs = prediction_layers(x)
model = Model(inputs,outputs)
```

```python
model.compile(optimizer = Adam(), loss = CategoricalCrossentropy(), metrics = ['accuracy'])
```

```python
model.history = model.fit(x=X_train, y=y_train, validation_data=(X_val, y_val), epochs=100)
```
